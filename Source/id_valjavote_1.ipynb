```python
import json
import folium
```

```python
# .json faili import 
# scrape_trt.json faili asemele peab panema json faili kuhu andmed salvestati.
with open('scrape_trt.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
# Asukoha punktide välja korjamine
locations = []
for marker in data.get('markers', []):
    try:
        lat = marker["0"]
        lon = marker["1"]
        locations.append((lat, lon))
    except KeyError:
        continue

print(f"välja korjati {len(locations)} punkti.")
```

```python
# Kaardi tegemine andmete kontrollimseks. pole otseselt vajalik.
# Näitab hästi aga kuidas andmeid klasterdati
map_center = [58.3776, 26.7290]
rental_map = folium.Map(location=map_center, zoom_start=13)
```

```python
for lat, lon in locations:
    folium.Marker([lat, lon]).add_to(rental_map)
```

```python
rental_map = folium.Map(location=[58.3776, 26.7290], zoom_start=13)

for marker in data["markers"]:
    lat = marker.get("0")
    lon = marker.get("1")
    count = marker.get("count", 1)
    object_ids = marker.get("object_ids", "")

    if lat is None or lon is None:
        continue

    ids = object_ids.split(",") if object_ids else []

    if count == 1 or not ids:
        # Single listing or unknown breakdown
        folium.Marker([lat, lon], popup="1 listing").add_to(rental_map)
    else:
        # Multiple listings at one location
        popup_text = f"{count} listings:\n" + "\n".join(ids)
        folium.Marker([lat, lon], popup=popup_text).add_to(rental_map)

rental_map
```

```python
# ID-d välja korjamine klastritest
all_ids = set()

for marker in data.get("markers", []):
    object_ids = marker.get("object_ids", "")
    if object_ids:
        all_ids.update(object_ids.split(","))

object_id_list = sorted(all_ids)

print(f"Välja korjati {len(object_id_list)} unikaalset object ID-d.")
```
